```python
#Easy Filter
actor_name = 'Sam Worthington'
matching_movies = movies[movies['actors'].apply(lambda x: actor_name in x)]
print(matching_movies[['title_y', 'actors']])

matching_movies
```

```python
movies.dropna(subset=['genre_names'], inplace=True)

# Filter movies based on actor and genre
actor_name = 'Sam Worthington'
#That are both Action & Thriller
genre1 = 'Action'
genre2 = "Thriller"
matching_movies = movies[movies['actors'].apply(lambda x: actor_name in x) & movies['genre_names'].apply(lambda x: all(genre in x for genre in [genre1, genre2]))]
matching_movies = matching_movies.sort_values(by='vote_average', ascending=False)

print(matching_movies[['title_y', 'actors', 'genre_names', "vote_average"]])
```

```python
def search_movies():
    # Prompt user for input
    actor_name = input("Enter actor's name: ")
    genre = input("Enter genre: ")

    # Filter movies
    matching_movies = movies[(movies['actors'].apply(lambda x: actor_name in x)) & (movies['genre_names'].apply(lambda x: all(g in x for g in genre.split(",")))) & (movies['genre_names'].notnull())]
    matching_movies = matching_movies.sort_values(by='vote_average', ascending=False)

    # Print matching movies
    print(matching_movies[['title_y', 'actors', 'genre_names', 'vote_average']])

# Call the function to start the search
search_movies()

```
